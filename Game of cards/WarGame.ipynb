```python
import random
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return self.rank + " of suit " + self.suit
```

```python
class Deck:
    def __init__(self):
        self.all_cards=[]
        for s in suits:
            for r in ranks:
                self.all_cards.append(Card(s,r))
    
    def shuffle_deck(self):
        random.shuffle(self.all_cards)
    
    def deal_from_deck(self): 
        return self.all_cards.pop() 
```

```python
class Player:
    def __init__(self,name):
        self.all_cards=[]
        self.name = name
        
    def grab_card(self,cards_l):
        if type(cards_l) == list :
            self.all_cards.extend(cards_l)
        else:
            self.all_cards.append(cards_l)
            
            
    def play_card(self):
        return self.all_cards.pop(0)
    
    def __str__(self):
        return f"Player {self.name} has {len(self.all_cards)} cards."
```

```python
def evaluate_game(player1, player2):
    if len(player1.all_cards)==0 and len(player2.all_cards)==0:
        print("The players tied")    
        return False 
        
    elif len(player1.all_cards)==0:
        print(f"Player {player1.name} wins!")    
        return False
    
    elif len(player2.all_cards)==0:
        print(f"Player {player2.name} wins!")    
        return False
    else:
        return True
```

```python
### War Game Logic

## Assigning names to players
player1_name = input("Please provide name of first player: ")
player2_name = input("Please provide name of second player: ")

player1 = Player(player1_name)
player2 = Player(player2_name)

## Create deck and shuffle cards

play_deck = Deck()
play_deck.shuffle_deck()


## Deal all cards to players

for _ in range(1,27):
    player1.grab_card(play_deck.deal_from_deck())
    player2.grab_card(play_deck.deal_from_deck())
    

play_round = 0
## Players start game
game_on = True

while game_on == True:
    table_player1=[] # Cards on the table from player1
    table_player2=[] # Cards on the table from player2
    
    table_player1.append(player1.play_card()) #Player 1 plays a card
    table_player2.append(player2.play_card()) # Play 2 plays a card
    
    while table_player1[-1].rank == table_player2[-1].rank: # Statement checks if war begins
        
        try: 
            table_player1.append(player1.play_card())
            table_player1.append(player1.play_card())
            table_player1.append(player1.play_card())
            table_player1.append(player1.play_card())
        except:
            pass
        try:
            table_player2.append(player2.play_card())
            table_player2.append(player2.play_card())
            table_player2.append(player2.play_card())
            table_player2.append(player2.play_card())
        except:
            pass
        
        
        
        
   ## Print carts in war     
    if len(table_player1)>1 or len(table_player2)>1:
        print("War!")
        count=0 # Counter of how many cards player 1 played in this battle 
        print("Player 1 ---------")
        for i in table_player1:
            if count%4==0 or count==len(table_player1)-1:
                print(i)
            else:
                print("Hidden Card")
            count+=1
        
        print("Player 2 ---------")
        count=0 # Counter of how many cards player 2 played in this battle 
        for i in table_player2:
            if count%4==0 or count==len(table_player2)-1:
                print(i)
            else:
                print("Hidden Card")
            count+=1
        if table_player1[-1].value > table_player2[-1].value:
            print(f"!!Player {player1.name} wins the battle!!\n")
        else:
            print(f"!!Player {player2.name} wins the battle!!\n")
            
    ##Evaluation of who gets the cards after this rund
    if game_on == True:    
        if table_player1[-1].value > table_player2[-1].value: 
            player1.grab_card(table_player1)
            player1.grab_card(table_player2)
            
                
        else:
            player2.grab_card(table_player1)
            player2.grab_card(table_player2)
            

        game_on = evaluate_game(player1, player2)
    
    ##Print results after 10 rounds
    play_round +=1
    if play_round % 10 == 0:
        print(f"---Results after {play_round} rounds---")
        print(f"Player {player1.name} || Cards: {len(player1.all_cards)}")
        print(f"Player {player2.name} || Cards: {len(player2.all_cards)}\n")
        
    ##Cards shuffle for both players
    random.shuffle(player1.all_cards)
    random.shuffle(player2.all_cards)
    
        
        
```
