```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1}
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return self.rank + " of suit " + self.suit
```

```python
class Deck:
    def __init__(self):
        self.all_cards=[]
        for s in suits:
            for r in ranks:
                self.all_cards.append(Card(s,r))
    
    def shuffle_deck(self):
        random.shuffle(self.all_cards)
    
    def deal_from_deck(self): 
        return self.all_cards.pop() 
```

```python
class Player:
    def __init__(self,name):
        self.all_cards = []
        self.name = name
        self.value = 0
        self.balance = 0
        
    def grab_card(self,cards_l):
        if type(cards_l) == list :
            self.all_cards.extend(cards_l)
        else:
            self.all_cards.append(cards_l)
        
        self.value = 0
        for card in self.all_cards:
            if card.rank =="Ace" and self.value <= 10:
                self.value += 11
            else:
                self.value += card.value
                
    def draw_money(self,amount):
        if self.balance >= amount:
            self.balance -= amount
            return True
        else:
            print("You don't have such amount on your account.")
            print(f"Your balance is: ${self.balance}")
            return False
            
    def __str__(self):
        return f"Player {self.name} has {len(self.all_cards)} cards."
```

```python
def show_cards(player, dealer, final=False):
    print("------------------")
    print("|--Dealer cards--|")
    for card_n in range(len(dealer.all_cards)):
        if card_n==0 and final == False:
            print("|  Card Coverd  |")
        else:
            print("|  " + str(dealer.all_cards[card_n]) + "  |")
    print("|--Player cards--|")
    for card in player.all_cards:
        print("|  " + str(card) + "  |")
    print("------------------")
    return
```
